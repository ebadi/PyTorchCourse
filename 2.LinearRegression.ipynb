In simple linear regression the goal is to find the line (`y = mx + b`) that is closest to the all data points


```python
# regression: the line that closest fit to the points
# pytorch uses automatic differntiation to find the grediants for simple linear ones
# but for complex models gradient descent is used
# learning rate (used in gradient descen) is a hyper parameters (used to train model but not a parameter to the model like w and b)
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 200
wx = 0.5
# random number between -5, +5
X = np.random.random(N) * 10 -5
Y = wx * X  + np.random.randn(N)
```

```python
plt.scatter(X, Y)
```

```python
model = nn.Linear(1,1)
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
type(X)
X= X.reshape(N,1)
Y= Y.reshape(N,1)
type(X)
```

```python
inputs = torch.from_numpy(X.astype(np.float32))
targets = torch.from_numpy(Y.astype(np.float32))
```

```python
type(inputs)
```

```python
n_epochs = 30
losses = []
for it in range(n_epochs):
  optimizer.zero_grad() # ??
  outputs = model(inputs) # forward pass
  loss = criterion(outputs, targets)
  losses.append(loss.item())
  loss.backward() # backward and optimize
  optimizer.step()
  print(f'Epoch {it+1}/{n_epochs}, Loss {loss.item():.4f}')
```

```python
plt.plot(losses);
```

```python
predicted = model(inputs).detach().numpy()
plt.scatter(X, Y, label="original data")
plt.plot(X, predicted, label= "fited line")
plt.legend()
plt.show()
```
