```python
!pip install geopy osmnx geopandas shapely
import osmnx as ox
import geopandas as gpd
from shapely.geometry import Point

def download_osm_data(city_name):
    # Download the street network data for the given city
    G = ox.graph_from_place(city_name, network_type='all')
    # Convert the graph to a GeoDataFrame
    nodes, edges = ox.graph_to_gdfs(G)
    return nodes, edges

def get_gps_coordinates(city_name, street_name, nodes, edges):
    # Perform the geocoding by searching for the street name in the edges GeoDataFrame
    streets = edges[edges['name'].str.contains(street_name, case=False, na=False)]

    if streets.empty:
        return "Street not found"

    # Get the first match
    street = streets.iloc[0]
    # Get the geometry of the street (LineString) and find its centroid
    point = street.geometry.centroid
    return (point.y, point.x)
nodes, edges = download_osm_data("Göteborgs")
```

```python
coordinates = get_gps_coordinates("Göteborgs", "Åvägen", nodes, edges)
print(coordinates)
```

```python
def gbgcoord(street):
  return get_gps_coordinates("Göteborgs", street, nodes, edges)


tmp = onlygbg
tmp['coordinates'] = tmp['street'].apply(gbgcoord)
```
